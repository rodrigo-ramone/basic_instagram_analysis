```python
# Convertendo a coluna hastags em uma lista como uma nova coluna
# Convert hashtags to list as a new column
data['Hashtags_List'] = data['Hashtags'].str.lower().str.strip().str.split()

# Adicionando coluna indicando se h√° hastags duplicadas na mesma postagem
# Adding a new column indicating if there are duplicated hashtags inside the same post
data['Has_Duplicate_Hashtags'] = data['Hashtags_List'].apply(lambda x: len(x) != len(set(x)))

# Criando nova coluna contendo quais hashtags est√£o repetidas
# Creating a new column having which hashtags are repeated
data['Repeated_Hashtags'] = data['Hashtags_List'].apply(
    lambda x: list({tag for tag in x if x.count(tag) > 1})
)
```

### üîç Passos Adicionais | Additional steps

‚ö†Ô∏è**Cria√ß√£o de vari√°veis de engagamento:** Gerando fun√ß√£o que passa pelas vari√°veis desejadas e cria vari√°veis com dois aspectos distintos.

* **Primeiro**, as vari√°veis desejadas s√£o dividas pelo **n√∫mero de impress√µes/visualiza√ß√µes** para obter m√©trica do quanto **engagamento** h√° **por visualiza√ß√£o**.

* **Valores altos indicam** que determinada vari√°vel tem **melhor convers√£o** (valores altos) por impress√£o/visualiza√ß√£o

* **Segundo**, o n√∫mero de **impress√µes/visualiza√ß√µes** √© dividido pelas vari√°veis desejadas para obter m√©trica do quanto de **esfor√ßo** √© **necess√°rio para aumentar a vari√°veis desejada**

* **Valores baixos indicam** que determinada vari√°vel tem **melhor desempenho**, pois necessita de menos impress√µes para aumentar seus valores

‚ö†Ô∏è**Engineering engagement features:** Creating a function that goes through the desired variables and create variables with two distinct aspects.

*  **First**, the desired **variables are divided by impressions** to create a metric of **how much engagement** there is **per impression**.

* **High values indicates** that a determined variable has **better conversion** (high values) per impression

*  **Second**, **number of impressions is divided by the desired variables** to create a metric of **how much effort** is **necessary to increase the desired variable**

* **Low values indicates** that a determined variable has **better performance** because it needs less impressions to increase its value


```python
def create_engagement_features(df, metrics, base_col="Impressions"):
    for col in metrics:
        rate_name = f"{col.lower()}_rate"
        df[rate_name] = df[col] / df[base_col]
        inverse_rate_name = f"impressions_per_{col.lower()}"
        df[inverse_rate_name] = df[base_col] / df[col].replace(0, np.nan)
    return df

metrics = ["Likes", "Comments", "Saves", "Follows", "Shares"]

unique_df_grouped = create_engagement_features(unique_df_grouped, metrics)
```

```python
# Ordenando os dados por follows_rate descrecente e impressions_per_follows ascendente
# Ordering data by descending follows_rate and ascending impressions_per_follows
unique_df_grouped.sort_values(by=['follows_rate', 'impressions_per_follows'], ascending=[False, True], inplace=True)
```

```python
# Removendo linhas com valores nulos
# Dropping rows that have Null values
unique_df_grouped = unique_df_grouped.dropna().copy()
```

### üîç An√°lise de Cluster | Clustering Analysis

**Observa√ß√£o:** Aqui vamos usar utilizar o m√©todo de matriz de TF-IDF que considera o qu√£o frequente e o qu√£o raro o termo √©, neste caso, cada hashtag. Aqui vamos utilizar algumas vari√°veis originais, al√©m de follows_rate e impressions_per_follows, j√° que a nossa vari√°vel de interesse √© follows

**Observation:** Here we will use the TF-IDF matriz method that considers how frequent and rare a term is, in this case each hashtag. Here, we will use some of the original features, besides follows_rate and impressions_per_follows, since follows is our target feature

```python
# Transformando cada hashtag em vetor num√©rico, removendo palavras sem import√¢ncia e garantindo que elas n√£o sejam comum demais
# Transforming each hashtag into a numeric vector, removing non-important words and making sure they are not too common
vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7, min_df=1)
X = vectorizer.fit_transform(unique_df_grouped['Hashtags'])
```

```python
# Separando vari√°veis a serem consideradas para a clusteriza√ß√£o
# Retrieving features to be considered for clustering
X = unique_df_grouped[['Likes', 'Comments', 'Saves', 'Impressions','Shares', 'follows_rate', 'impressions_per_follows']]

# padronizando as vari√°veis
# Standardizing features
X_scaled = StandardScaler().fit_transform(X)

# Usando o m√©todo Elbow para identificar o n√∫mero de clusters mais apropriado
# Using the Elbow method to identify the appropriate number of clusters
WCSS = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000, random_state=1)
    kmeans.fit(X_scaled)
    WCSS[k] = kmeans.inertia_

# Plotando o gr√°fico Elbow
# Plotting the Elbow graph
plt.figure()
plt.plot(list(WCSS.keys()), list(WCSS.values()), 'bx-')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method for Hashtag Clustering")
plt.show()
```

```python
# Determinando o n√∫mero de clusters e aplicando m√©todo de kmeans para separar as hashtags em clusters
# Setting the number of clusters and applying the kmeans method to separate hashtags into clusters
k = 2
kmeans = KMeans(n_clusters=k, random_state=42)
unique_df_grouped['Hashtag_Cluster'] = kmeans.fit_predict(X)
```

```python
# Transformando vari√°veis n√£o num√©ricas em vari√°veis dummy
# Transforming non-numerical variables in dummy variables
unique_df_dummy = pd.get_dummies(
    unique_df_grouped,
    columns=['Hashtag_Cluster'],
    drop_first = True,
)
```

```python
# Plotando gr√°ficos para mostrar as diferen√ßas entre os dois clusters
# Ploting graphs to show the diffences between the two clusters
fig, axes = plt.subplots(1, 6, figsize=(18, 5))

# Plot 1: Hashtag Count vs Impressions
sns.boxplot(data=unique_df_dummy, x='Hashtag_Cluster_1',y='Follows', ax=axes[0])
axes[0].set_title('Hashtag Cluster vs Follows')

# Plot 2: Hashtag Count vs Likes
sns.boxplot(data=unique_df_dummy, x='Hashtag_Cluster_1',y='Likes', ax=axes[1])
axes[1].set_title('Hashtag Cluster vs Likes')

# Plot 3: Hashtag Count vs Follows
sns.boxplot(data=unique_df_dummy, x='Hashtag_Cluster_1',y='Impressions', ax=axes[2])
axes[2].set_title('Hashtag Cluster vs Impressions')

# Plot 4: Hashtag Count vs Follows
sns.boxplot(data=unique_df_dummy, x='Hashtag_Cluster_1',y='Comments', ax=axes[3])
axes[3].set_title('Hashtag Cluster vs Comments')

# Plot 5: Hashtag Count vs Follows
sns.boxplot(data=unique_df_dummy, x='Hashtag_Cluster_1',y='Shares', ax=axes[4])
axes[4].set_title('Hashtag Cluster vs Shares')

# Plot 6: Hashtag Count vs Follows
sns.boxplot(data=unique_df_dummy, x='Hashtag_Cluster_1',y='follows_rate', ax=axes[5])
axes[5].set_title('Hashtag Cluster vs Follows Rate')

plt.tight_layout()
plt.show()
```

**Modelagem e simula√ß√£o:** Agora iremos criar nosso modelo para simular como se comporta Follows e Follows_rate se as 10 hashtags n√£o importantes tivessem o mesmo n√∫mero de visualiza√ß√µes das 10 hashtags mais importantes

**Modelling and simulation:** Now we will create our model to simulate how Follows and Follows_rate would behave if the top 10 non-important hashtags had the same number of impressions as the top 10 important hashtags

```python
# Selecionar top 10 non_important (ordenado por Follows)
# Select top 10 non-important (ordered by Follows)
top_non_important = non_important_hashtags.sort_values('Follows', ascending=False).head(10).reset_index(drop=True)
top_important = important_hashtags.sort_values('Follows', ascending=False).head(10).reset_index(drop=True)

# C√≥pia para n√£o alterar original
# Copy to not alter the original
top_non_important_sim = top_non_important.copy()

# Impress√µes simuladas (das hashtags importantes)
# Simulated impressions (From important hashtags)
top_non_important_sim['Impressions_sim'] = top_important['Impressions']

# Calcula Follows simulados mantendo o Follows_rate original
# Calculate simulated Follows keeping the original Follows_rate
top_non_important_sim['Follows_sim'] = top_non_important_sim['follows_rate'] * top_non_important_sim['Impressions_sim']

# Recalcula Follows_rate simulado
# Recalculate the simulated Follows_rate
top_non_important_sim['Follows_rate_sim'] = top_non_important_sim['Follows_sim'] / top_non_important_sim['Impressions_sim']

# Ganho percentual de follows
# Porcentual follows gain
top_non_important_sim['Follows_Gain_%'] = ((top_non_important_sim['Follows_sim'] - top_non_important_sim['Follows']) / top_non_important_sim['Follows']) * 100

# Resultado final
# Final result
top_non_important_sim
```

**Observa√ß√µes:** Aqui percebemos que mantendo o padr√£o de follows_rate mas tendo as mesmas visualiza√ß√µes que as top 10 hashtags importantes, as top 10 hashtags n√£o importantes aumentam consideravelmente o n√∫mero de seguidores (Follows_sim), o que √© confirmado pela coluna Follows_gain, que indica a mudan√ßa em porentagem em rela√ß√£o ao n√∫mero de seguidores com visualiza√ß√µes originais

**Observations:** Here, we notice that keeping the follows_rate pattern but having the number of impressions from top 10 important hashtags, the top 10 non-important hashtags increase followers considerably (Follows_sim), that is confirmed by Follows_gain which indicates the porcentage change of followers in relation to the original number of impressions

**√öltimo passo:** Agora iremos calcular se as top 10 hashtags importantes continuam com mais seguidores do que as top 10 hashtags n√£o importantes

**Last step:** Now we will calculate if the top 10 important hashtags keep on having more followers than the top 10 non-important hashtags

```python
# Soma de follows das top 10 important hashtags
# Follows Sum of top 10 important hashtags
original_follows_sum = top_important['Follows'].sum()

# Soma de follows simulados para as top 10 hashtags n√£o importantes
# Simulated follows sum for the top 10 non-important hashtags
predicted_follows_sum = top_non_important_sim['Follows_sim'].sum()

# Diferen√ßa absoluta e percentual
# Absolute and percentage difference
diff_follows = predicted_follows_sum - original_follows_sum
percent_diff_follows = (diff_follows / original_follows_sum) * 100

print(f"Difference in follows / Diferen√ßa de seguidores: {diff_follows:.0f}")
print(f"Percentage difference in follows / Diferen√ßa de seguidores em porcentagem: {percent_diff_follows:.2f}%")

# Compara√ß√£o entre a raz√£o de seguidores agregada das hashtags importantes e das n√£o importantes com valores simulados
# Agregated follows rate comparison between top important hashtags and top non-important with simulated values
original_rate_sum = top_important['follows_rate'].sum()
predicted_rate_sum = top_non_important_sim['Follows_rate_sim'].sum()

diff_rate = predicted_rate_sum - original_rate_sum
percent_diff_rate = (diff_rate / original_rate_sum) * 100

print(f"Difference in follows_rate / Diferen√ßa na taxa: {diff_rate:.4f}")
print(f"Percentage difference in follows_rate / Diferen√ßa na taxa em porcentagem: {percent_diff_rate:.2f}%")
```
